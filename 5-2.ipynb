```python
from sklearn.datasets import load_wine
```

```python
wine = load_wine()
X = wine.data
y = wine.target
```

```python
print("Wine dataset size:",X.shape)
```

# Applying PCA

```python
from sklearn.preprocessing import StandardScaler 
X_scaled = StandardScaler().fit_transform(X)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

# Making the Scatter plot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[7,5])
plt.scatter(X_pca[:,0], X_pca[:,1], c=y, s=25, cmap='plasma')
plt.title("PCA of wine with 2 compo.")
plt.xlabel("Comp 1")
plt.ylabel("Comp 2")
```

```python
exp_var = sum(pca.explained_variance_ratio_ * 100)
print('Varience Explained', exp_var)
```

# Applying LDA 

```python
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=2)
X_lda = lda.fit_transform(X_scaled, y)
```

# Making Scatterplot

```python
import matplotlib.pyplot as plt
plt.figure(figsize=[7,5])
plt.scatter(X_lda[:,0], X_lda[:,1], c=y, s=25, cmap="plasma")
plt.xlabel('Comp 1')
plt.ylabel('Comp 2')
plt.title('LDA scatterplot')
```

```python
exp_var = sum(lda.explained_variance_ratio_ *100)
print("Explained Variance:", exp_var)
```

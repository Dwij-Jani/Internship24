```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("counterfeit_train.csv")
data
```

```python
data['Counterfeit_Weight'] = data['Counterfeit_Weight'].fillna(data['Counterfeit_Weight'].mean())
data.isnull().sum()
```

```python
clm_drop = data[['Medicine_ID','DistArea_ID']]
data = data.drop(clm_drop,axis=1)
data
```

```python
data_dumm = pd.get_dummies(data[['Medicine_Type','SidEffect_Level','SidEffect_Level','Area_Type','Area_City_Type','Area_dist_level']])
data_dumm = data_dumm.astype(float)
```

```python
data = pd.concat([data, data_dumm],axis=1)
data = data.drop(['Medicine_Type','SidEffect_Level','SidEffect_Level','Area_Type','Area_City_Type','Area_dist_level'],axis=1)
```

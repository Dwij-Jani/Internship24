```python
import numpy as np
import pandas as pd
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
df = pd.read_csv('counterfeit_train.csv') 
df
```

```python
df.isnull().sum()
```

```python
df['Counterfeit_Weight'] = df['Counterfeit_Weight'].fillna(df['Counterfeit_Weight'].mean())
df
```

```python
df= df.drop(['Medicine_ID','DistArea_ID'],axis=1)
```

```python
cat_values = df.select_dtypes(include=['object'])
cat_values.head()
```

```python
data_dummies = pd.get_dummies(cat_values, drop_first=True)
data_dummies.head()
data_dummies = data_dummies.replace({True:1, False:0})
```

```python
data_dummies
```

```python
x = df.iloc[::,:4]
y = df.iloc[::,-1]
```

```python
df
```

```python
y
```

```python
model = Sequential()
model.add(Dense(4, input_dim = 4, activation='relu'))
model.add(Dense(4, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='mean_squared_error',optimizer='adam', metrics=['mse','mae'])
model.fit(x, y, epochs=30, batch_size =15)
```
